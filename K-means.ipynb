```python
df=df.drop(columns=['ADDRESSLINE1', 'ADDRESSLINE2', 'STATE', 'POSTALCODE'])
```

```python
df.isnull().sum()
```

```python
import sklearn
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
scaled_data = sc.fit_transform(df[['QUANTITYORDERED', 'SALES']])
```

```python
scaled_data
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    kmeans= KMeans(n_clusters=i)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss,marker='o')
```

```python
optimal_k=3
```

```python
kmeans= KMeans(n_clusters=optimal_k)
```

```python
df['Clusters']=kmeans.fit_predict(scaled_data)
```

```python
df
```

```python
import seaborn as sns
for i in range(optimal_k):
  plt.scatter(df[df['Clusters']==i]['QUANTITYORDERED'],df[df['Clusters']==i]['SALES'])
plt.legend()
plt.show()
```

```python
sns.scatterplot(data=df,x='QUANTITYORDERED',y='SALES',hue='Clusters')
```
